## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### üí° Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
Build a budgeting program that allows the user to:

Allows users to input their monthly income
- Allows users to input up to 3 expenses
- Calculates remaining budget
- Shows a warning if budget is negative
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Get user's name
name = input("Please enter your name: ")

# Get user's monthly income with validation
while True:
    try:
        income = float(input("Please enter your monthly income: "))
        break
    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Print personalized welcome message
print(f"\nWelcome, {name}!")
print(f"Your monthly income is: {income}")

# Initialize total expenses
total_expenses = 0

# Get up to 3 expenses
print("\nPlease enter up to 3 expenses:")
for i in range(1, 4):
    while True:
        try:
            expense = float(input(f"Enter expense {i}: "))
            total_expenses += expense
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n" + "="*40)
print("BUDGET SUMMARY")
print("="*40)
print(f"Monthly Income: ${income:.2f}")
print(f"Total Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")
print("="*40)

# Show warning if budget is negative
if remaining_budget < 0:
    print("\n‚ö†Ô∏è  WARNING: Your budget is NEGATIVE!")
    print(f"You are over budget by ${abs(remaining_budget):.2f}")
else:
    print(f"\n‚úì You have ${remaining_budget:.2f} remaining in your budget.")

print("\nThank you for using the budgeting program!")

# [Paste your code here]
```

```python
# REFINED VERSION - Your Working Budget Tracker
# Get user's name
name = input("Please enter your name: ")

# Get user's monthly income with validation
while True:
    try:
        income = float(input("Please enter your monthly income: "))
        break
    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Print personalized welcome message
print(f"\nWelcome, {name}!")
print(f"Your monthly income is: {income}")

# Initialize total expenses and expense list
total_expenses = 0
expenses = []

# Get up to 3 expenses
print("\nPlease enter up to 3 monthly expenses:")
for i in range(1, 4):
    expense_name = input(f"\nEnter expense {i} name: ")

    while True:
        try:
            expense_amount = float(input(f"Enter expense {i} amount: "))
            expenses.append((expense_name, expense_amount))
            total_expenses += expense_amount
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n" + "="*40)
print("BUDGET SUMMARY")
print("="*40)
print(f"Monthly Income: ${income:.2f}")
print("\nExpenses:")
for expense_name, expense_amount in expenses:
    print(f"  - {expense_name}: ${expense_amount:.2f}")
print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")
print("="*40)

# Show warning if budget is negative
if remaining_budget < 0:
    print("\n‚ö†Ô∏è  WARNING: Your budget is NEGATIVE!")
    print(f"You are over budget by ${abs(remaining_budget):.2f}")
else:
    print(f"\n‚úì You have ${remaining_budget:.2f} remaining in your budget.")

print("\nThank you for using the budgeting program!")
# [Paste your corrected/improved code here]
```

```python
# ENHANCED BUDGET TRACKER
# Get user's name
name = input("Please enter your name: ")

# Get user's monthly income with validation
while True:
    try:
        income = float(input("Please enter your monthly income: "))
        if income < 0:
            print("Invalid input. Income cannot be negative.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Print personalized welcome message
print(f"\nWelcome, {name}!")
print(f"Your monthly income is: {income}")

# Initialize total expenses and expense list
total_expenses = 0
expenses = []

# Get up to 3 expenses
print("\nPlease enter up to 3 expenses:")
for i in range(1, 4):
    expense_name = input(f"\nEnter expense {i} name: ")

    while True:
        try:
            expense_amount = float(input(f"Enter expense {i} amount: "))
            if expense_amount < 0:
                print("Invalid input. Expense amount cannot be negative.")
            else:
                expenses.append((expense_name, expense_amount))
                total_expenses += expense_amount
                break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n" + "="*40)
print("BUDGET SUMMARY")
print("="*40)
print(f"Monthly Income: ${income:.2f}")
print("\nExpenses:")
for expense_name, expense_amount in expenses:
    print(f"  - {expense_name}: ${expense_amount:.2f}")
print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")
print("="*40)

# Show warning if budget is negative
if remaining_budget < 0:
    print("\n‚ö†Ô∏è  WARNING: Your budget is NEGATIVE!")
    print(f"You are over budget by ${abs(remaining_budget):.2f}")
else:
    print(f"\n‚úì You have ${remaining_budget:.2f} remaining in your budget.")

print("\nThank you for using the budgeting program!")
# [Paste your enhanced code here]
```

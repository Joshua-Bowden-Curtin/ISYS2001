## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
amount = float(input("Enter expense amount: $"))

if amount > 100:
    category = "Major expense"
elif amount > 50:
    category = "Moderate expense"
elif amount > 10:
    category = "Minor expense"
else:
    category = "Small purchase"

print(f"${amount:.2f} → {category}")
````

````python
#print("=== Budget Limit Checker ===")

# Get user's budget and expense
#monthly_budget = float(input("Enter your monthly budget: $"))
#expense = float(input("Enter expense amount: $"))

print("=== Budget Limit Checker ===")

expenses = []

# Get user's monthly budget
monthly_budget = float(input("Enter your monthly budget: $"))

while True:
    entry = input("Enter expense amount (or 'Done' to finish): $")

    if entry.lower() == "done":
        break

    expense = float(entry)

    # Decision logic
    if expense > monthly_budget * 0.50:
        category = "Major purchase - think carefully"
    elif expense > monthly_budget * 0.25:
        category = "Significant expense"
    elif expense > monthly_budget * 0.10:
        category = "Moderate expense"
    else:
        category = "Minor expense - within budget"

    expenses.append({"amount": expense, "category": category})
    print(f"  → {category}")

# Summary
print("\n=== Expense Summary ===")
print(f"Monthly budget: ${monthly_budget:.2f}")
for expense in expenses:
    print(f"  ${expense['amount']:.2f} — {expense['category']}")

total = sum(e["amount"] for e in expenses)
print(f"\nTotal spent:    ${total:.2f}")
print(f"Remaining:      ${monthly_budget - total:.2f}")

# Decision logic here - you write this part!
# If expense > 50% of budget: "Major purchase - think carefully"
# If expense > 25% of budget: "Significant expense"
# If expense > 10% of budget: "Moderate expense"
# Otherwise: "Minor expense - within budget"
````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python
#\print("=== Savings Goal Tracker ===")

#goal_amount = float(input("Savings goal: $"))
#current_savings = float(input("Current savings: $"))

# Your task: Calculate percentage and give encouraging feedback
# 100%+: "Goal achieved! Consider a new goal."
# 75-99%: "Almost there! You're doing great."
# 50-74%: "Halfway there - keep it up!"
# 25-49%: "Good progress, stay consistent"
# Under 25%: "Just getting started - you've got this!"

print("=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

while True:
    entry = input("Enter deposit amount (or 'Done' to finish): $")

    if entry.lower() == "done":
        break

    deposit = float(entry)
    current_savings += deposit

    percentage = (current_savings / goal_amount) * 100

    # Decision logic
    if percentage >= 100:
        feedback = "Goal achieved! Consider a new goal."
    elif percentage >= 75:
        feedback = "Almost there! You're doing great."
    elif percentage >= 50:
        feedback = "Halfway there - keep it up!"
    elif percentage >= 25:
        feedback = "Good progress, stay consistent."
    else:
        feedback = "Just getting started - you've got this!"

    print(f"  → ${current_savings:.2f} saved ({percentage:.1f}%) — {feedback}")

# Summary
print("\n=== Savings Summary ===")
print(f"Goal:           ${goal_amount:.2f}")
print(f"Total saved:    ${current_savings:.2f}")
print(f"Remaining:      ${max(goal_amount - current_savings, 0):.2f}")
print(f"Progress:       {min((current_savings / goal_amount) * 100, 100):.1f}%")
````

## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
#print("=== Smart Purchase Advisor ===")

# Get financial information
#available_money = float(input("How much money do you have available? $"))
#purchase_price = float(input("How much does the item cost? $"))
#monthly_income = float(input("What's your monthly income? $"))

# Calculate affordability percentage
#affordability_ratio = purchase_price / available_money

# Your task: Build the decision logic
# Consider:
# - Can they afford it at all?
# - Is it a large percentage of their available money?
# - Is it reasonable compared to their income?
# - What advice should you give?

# Suggestion: Use multiple if/elif conditions to give personalised advice

print("=== Smart Purchase Advisor ===")

while True:
    print("\n--- New Purchase Check ---")

    # Get financial information
    available_money = float(input("How much money do you have available? $"))
    purchase_price = float(input("How much does the item cost? $"))
    monthly_income = float(input("What's your monthly income? $"))

    # Calculate ratios
    affordability_ratio = purchase_price / available_money
    income_ratio = purchase_price / monthly_income

    print("\n  → Advisor says:")

    # 1. Basic affordability check
    if purchase_price > available_money:
        print("  ✗ You cannot afford this purchase.")
        print(f"    You are ${purchase_price - available_money:.2f} short.")

    else:
        # 2. How large a chunk of available money is it?
        if affordability_ratio > 0.75:
            print("  ⚠ This would use over 75% of your available money.")
            print("    Consider waiting until you have more saved up.")
        elif affordability_ratio > 0.50:
            print("  ⚠ This is a significant portion of your available funds.")
            print("    Make sure you keep enough aside for emergencies.")
        elif affordability_ratio > 0.25:
            print("  ✓ This is manageable, but will make a noticeable dent.")
            print("    Think about whether this is a need or a want.")
        else:
            print("  ✓ This looks comfortably affordable.")

        # 3. Check against monthly income
        if income_ratio > 1.0:
            print("  ⚠ This costs more than your entire monthly income — plan carefully.")
        elif income_ratio > 0.50:
            print("  ⚠ This is more than half your monthly income — a major commitment.")
        elif income_ratio > 0.25:
            print("  → This is a notable expense relative to your income.")
        else:
            print("  ✓ This is reasonable compared to your monthly income.")

        # 4. Overall recommendation
        print("\n  Recommendation:")
        if affordability_ratio <= 0.25 and income_ratio <= 0.25:
            print("  → Go for it — this purchase looks well within your means.")
        elif affordability_ratio <= 0.50 and income_ratio <= 0.50:
            print("  → Proceed with confidence, but keep an eye on your budget.")
        elif purchase_price <= available_money and income_ratio <= 1.0:
            print("  → You can afford it, but consider waiting and saving more first.")
        else:
            print("  → This purchase carries financial risk — strongly consider waiting.")

    # Summary figures
    print(f"\n  Purchase price:       ${purchase_price:.2f}")
    print(f"  Available money:      ${available_money:.2f}")
    print(f"  Left after purchase:  ${max(available_money - purchase_price, 0):.2f}")
    print(f"  % of available money: {affordability_ratio * 100:.1f}%")
    print(f"  % of monthly income:  {income_ratio * 100:.1f}%")

    # Continue or exit
    again = input("\nCheck another purchase? (yes / Done): ")
    if again.lower() == "done":
        print("\nThanks for using Smart Purchase Advisor. Spend wisely!")
        break
````

- **Reasonable purchase**: "This seems like a reasonable purchase for your budget"

## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
#print("=== Personal Finance Tracker v0.1 ===")
#print("Week 3: Smart Budget Decisions")
#print()

# User setup
#name = input("Enter your name: ")
#monthly_income = float(input("Enter your monthly income: $"))

#print(f"\nWelcome to your finance tracker, {name}!")
#print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
#print("\n--- Expense Entry ---")
#expense_amount = float(input("Enter an expense: $"))
#expense_description = input("What was this expense for? ")

# Apply your decision-making skills here
# Categorise the expense and give feedback
# This becomes the foundation you'll build on each week!

print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# Expense tracking
expenses = []
total_spent = 0

while True:
    print("\n--- Expense Entry ---")
    entry = input("Enter an expense (or 'Done' to finish): $")

    if entry.lower() == "done":
        break

    expense_amount = float(entry)
    expense_description = input("What was this expense for? ")

    # Budget percentage calculations
    budget_ratio = expense_amount / monthly_income
    total_spent += expense_amount
    total_ratio = total_spent / monthly_income

    # Expense categorisation
    if budget_ratio > 0.30:
        category = "Major expense"
        advice = "This is a large chunk of your income — was this planned?"
    elif budget_ratio > 0.15:
        category = "Significant expense"
        advice = "Notable spend. Make sure this fits your monthly plan."
    elif budget_ratio > 0.05:
        category = "Moderate expense"
        advice = "Reasonable spend — keep tracking to stay on top of things."
    else:
        category = "Minor expense"
        advice = "Small purchase — these can add up though, so keep an eye on them."

    # Budget health check based on cumulative spending
    if total_ratio > 0.90:
        budget_status = "⚠ Warning: you've used over 90% of your income this month!"
    elif total_ratio > 0.75:
        budget_status = "⚠ Caution: you're approaching your monthly limit."
    elif total_ratio > 0.50:
        budget_status = "→ You've used over half your monthly income so far."
    else:
        budget_status = "✓ Your spending is looking healthy so far."

    # Store the expense
    expenses.append({
        "description": expense_description,
        "amount": expense_amount,
        "category": category
    })

    # Feedback for this expense
    print(f"\n  {name}, here's your feedback:")
    print(f"  Category:    {category}")
    print(f"  This expense: {budget_ratio * 100:.1f}% of your monthly income")
    print(f"  Advice:      {advice}")
    print(f"\n  Total spent so far: ${total_spent:.2f} ({total_ratio * 100:.1f}% of income)")
    print(f"  {budget_status}")

# Final summary
print(f"\n{'=' * 40}")
print(f"  Monthly Summary for {name}")
print(f"{'=' * 40}")
print(f"  Monthly income:   ${monthly_income:.2f}")
print(f"  Total spent:      ${total_spent:.2f}")
print(f"  Remaining:        ${monthly_income - total_spent:.2f}")
print(f"  Budget used:      {(total_spent / monthly_income) * 100:.1f}%")
print(f"\n  Expenses breakdown:")
for e in expenses:
    print(f"    ${e['amount']:.2f}  —  {e['description']}  [{e['category']}]")
print(f"\n  See you next week, {name}!")
````
